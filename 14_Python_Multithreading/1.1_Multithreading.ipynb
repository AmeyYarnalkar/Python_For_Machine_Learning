```python
import threading
```

```python
# Define the task
def task():
    print("Task is executing")
    
# create the thread
t1 = threading.Thread(target=task)

print("Starting the task execution")
t1.start() # starts the thread execution
t1.join() # waits until thread t1 finishes its work
print("Done with execution")
```

```python
# Multiple thread creation
import time

import threading
import time

def task(parent_name):
    print(f"Executing task from {parent_name}")
    time.sleep(5)
    print(f"Done executing task from {parent_name}")

t1 = threading.Thread(target=task, args=("Thread-1",))
t2 = threading.Thread(target=task, args=("Thread-2",))

t1.start()
t2.start()

t1.join()
t2.join()
```

```python
# Race condition (Not necessary that it happens all the time)
counter = 0

def increment():
    global counter
    for _ in range(1000000):
        counter += 1
        
t1 = threading.Thread(target=increment)
t2 = threading.Thread(target=increment)
t3 = threading.Thread(target=increment)
t4 = threading.Thread(target=increment)

t1.start()
t2.start()
t3.start()
t4.start()

t1.join()
t2.join()
t3.join()
t4.join()

print(counter) 
```

```python
counter = 0
lock = threading.Lock()

def increment():
    global counter
    for _ in range(100000):

        lock.acquire()

        counter += 1

        lock.release()

t1 = threading.Thread(target=increment)
t2 = threading.Thread(target=increment)

t1.start()
t2.start()

t1.join()
t2.join()

print(counter)
```

```python
# Thread pull execution
from concurrent.futures import ThreadPoolExecutor

def task(n):
    return n ** n

with ThreadPoolExecutor(max_workers=4) as executor:
    result = executor.map(task, range(10))

print(list(result))
```
